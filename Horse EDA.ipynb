# Initial analysis

```python
SELECT Age, 
COUNT(Won) as Number_of_won_races, 
count(*) * 100.0 / sum(count(*)) over() as win_percentage
FROM horses
WHERE won = 1
GROUP BY Age
ORDER BY Age
```

```python
SELECT WeightValue, 
COUNT(Won) as Number_of_won_races, 
count(*) * 100.0 / sum(count(*)) over() as win_percentage
FROM horses
WHERE won = 1
GROUP BY WeightValue
ORDER BY WeightValue
```
